```python
#************************************************************

#3 Genetic Algorithm

import numpy as np
import matplotlib.pyplot as plt

# Fitness function to maximize
def fitness(x):
    return x * np.sin(x)

# GA settings
size = 20              # population size
generations = 50
cross_rate = 0.8
mutate_rate = 0.1
lower, upper = 0, 10   # x range

# Step 1: Random initial population
pop = np.random.uniform(lower, upper, size)

# Track best values
best_scores = []

for gen in range(generations):
    fit = fitness(pop)

    # Step 2: Keep best half
    top = pop[np.argsort(fit)[-size//2:]]

    # Step 3-5: Make new population
    new_pop = []
    while len(new_pop) < size:
        # Crossover
        if np.random.rand() < cross_rate:
            p1, p2 = np.random.choice(top, 2, replace=False)
            child = (p1 + p2) / 2
        else:
            child = np.random.choice(top)

        # Mutation
        if np.random.rand() < mutate_rate:
            child += np.random.uniform(-1, 1)

        # Keep in range
        child = np.clip(child, lower, upper)
        new_pop.append(child)

    pop = np.array(new_pop)
    best = np.max(fitness(pop))
    best_scores.append(best)

    print(f"Gen {gen+1}: Best = {best:.4f}")

# Final result
best_x = pop[np.argmax(fitness(pop))]
print(f"\nBest x = {best_x:.4f}, f(x) = {fitness(best_x):.4f}")

# Plot result
plt.plot(best_scores)
plt.title("Best Fitness Over Generations")
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.grid()
plt.show()

#************************************************************
```

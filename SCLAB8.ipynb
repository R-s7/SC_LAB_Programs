```python
#************************************************************

#8 Crisp Partition For Real Life Iris Dataset

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Step 2: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply KMeans clustering (crisp partitioning)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Step 4: Create a DataFrame for analysis
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['True Class'] = [target_names[i] for i in y]
df['Cluster'] = clusters

# Step 5: Print cluster assignments and compare with true labels
print("Sample data with cluster assignments:")
print(df.head())

print("\nCluster distribution (True Class vs Cluster):")
print(df[['True Class', 'Cluster']].value_counts())

# Step 6: Plot the clusters (using first two features for visualization)
plt.figure(figsize=(10, 6))
colors = ['red', 'green', 'blue']
for i in range(3):
    plt.scatter(X_scaled[df['Cluster'] == i, 0],
                X_scaled[df['Cluster'] == i, 1],
                c=colors[i], label=f'Cluster {i}')

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('K-Means Clustering (Crisp Partition) - Iris Dataset')
plt.legend()
plt.grid(True)
plt.show()

#************************************************************
```
